## VI. Pendefinisian Model

Selanjutnya kita bisa melakukan persiapan datanya

```python
datafix_ML = a[['distance','price']]
```

```python
datafix_ML
```

Kita reshape supaya dapat menjadi suatu matriks yang nanti kita visualisasikan

```python
x = datafix_ML['distance'].values.reshape(-1,1)
y = datafix_ML['price'].values.reshape(-1,1)
```

```python
#selanjutnya dapat kita split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.8, random_state=100)
```

```python
model_ML = LinearRegression(fit_intercept=True)
model_ML.fit(x_train, y_train)
```

Selanjutnya dapat kita cari slope dan biasnya dengan cara sebagai berikut

```python
print("Koef Model Machine Learningnya adalah :", model_ML.coef_)
print("Bias dari persamaannya adalah :", model_ML.intercept_)
y_predict = model_ML.predict(x_test)
```

dari data tersebut kita bisa melihat bahwa variable didepan x nya adalah 2 sehingga disini didapatkan gradiennya adalah 2, dan juga didapatkan suatu koefisien biasnya adalah 6 

```python
plt.xlabel('distance')
plt.ylabel('price')
x = np.linspace(0, 10, 1000)
y = model_ML.predict(x[:, np.newaxis])
plt.scatter(x_train, y_train)
plt.plot(x, y, 'r-');
plt.title('Lytf', fontdict={'fontsize' : 25})
```

## VII. Evaluasi Model

Kita dapat melihat dari nilai MSE nya dengan cara sebagai berikut ini

```python
print("Nilai dari MSE adalah :", mean_squared_error(y_test, y_predict))
```

dilihat dari data tersebut maka didapatkan hasil eror 2.4 sehingga dalam hal ini prediksi yang kita buat dapat memiliki standar deviasi atau kemungkinan error/meleset sebesar 2.4 dollar
